+ Standardize the column names making use of camel_case

+ Select only books with at least 1 Rating;

#### Data Cleaning 

```python
#Definiting a function to clean the column names 
def edit_col(col):
    col = col.str.replace(" ","_")
    col = col.str.strip()
    col = col.str.replace("(","")
    col = col.str.replace(".","")
    col = col.str.replace(")","")
    col = col.str.lower()
    return col
df.columns = edit_col(df.columns)
```

```python
#Cleaning authors column by removing the square brackets
df["authors"] = df["authors"].str.replace("[", "").str.replace("]","").str.replace("'","")

#Removing comma from Num. of Ratings 
df["num_of_ratings"] = df["num_of_ratings"].str.replace(",", "").astype(int)

# stardardizing all titles
df['title'] = df['title'].apply(lambda x: x.capitalize())

# dropping  duplicate rows
df.drop_duplicates(keep='last', inplace=True)

# selecting only rows with at least 1 Rating
df = df[df['num_of_ratings'] != 0]

# creating a new column for Weighted Avg to be used later
df['weighted_rating'] = df['avg_rating'] * df['num_of_ratings']

# creating BookID column
r1, r2 = 1, 990
df['BookID'] = np.arange(r1, r2+1, 1)
```

### 2.3 Exploring Author(s)

##### I first create a function that separate author from co-author(s) 

From which I will be able to calculate total numbers of books written by which author 

```python
def authors_df(author, feature):
    authors = author.map(lambda x: x.split(','))
    features = feature
    
    auth_feature = list(map(lambda tuple_auth_feature: 
                     list(map(lambda i: (i, tuple_auth_feature[1]), tuple_auth_feature[0])),
                     zip(authors, features)))
    
    flat_list = []
    for sublist in auth_feature:
        for item in sublist:
            flat_list.append(item)
    
    flat_list = set(flat_list)
    
    authors = list(map(lambda i: i[0], flat_list))
    features = list(map(lambda i: i[1], flat_list))
    df_auth_feature = pd.DataFrame({'Authors': authors, 'Feature': features})
    
    return df_auth_feature
```

##### Finding out authors who wrote the most books 

```python
# calling the function to create a new dataframe with the individual authors
most_books = authors_df(df['authors'], df['title'])


# selecting the top 20 by counting the books
top_20_most_books = pd.crosstab(index=most_books['Authors'],columns='count').sort_values(['count'], 
                                                            ascending=False).reset_index().head(20)
```

##### Finding most Popular authors 
This is determined by authors with the most number of ratings 

```python
most_popular_authors = authors_df(df['authors'], df['num_of_ratings'])
top_20_most_popular = most_popular_authors.groupby(['Authors'])['Feature'].sum().sort_values(
                            ascending=False).reset_index('Authors').drop([3], axis=0).head(20)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1, 2, 1)
ax1 = sns.barplot(x=top_20_most_books['count'], y=top_20_most_books['Authors'][2:], palette = 'hot')
plt.ylabel('')
plt.xlabel('Counts', fontsize=12)
plt.yticks(fontsize=12)
plt.title('\nAuthors Who Wrote Most Books\n', fontsize=15, fontweight='bold')
ax1.get_yticklabels()[0].set_color("r")
ax1.get_yticklabels()[1].set_color("r")
ax1.get_yticklabels()[2].set_color("r")


plt.subplot(1, 2, 2)
plt.subplots_adjust(wspace = 0.5)
ax2 = sns.barplot(x=top_20_most_popular['Feature'], y=top_20_most_popular['Authors'], palette = 'hot')
plt.ylabel('')
plt.xlabel('Number of Ratings', fontsize=12)
plt.yticks(fontsize=12)
plt.title('\nMost Popular Authors\n', fontsize=15, fontweight='bold')
ax2.get_yticklabels()[0].set_color("r");
ax2.get_yticklabels()[1].set_color("r")
ax2.get_yticklabels()[6].set_color("r")
```

The visualization above indicates that the number of books written by each author doesn't seem to much determine the popularity of the author as we can see in the case of <b> Richard Branson</b> who is the least writer among the top 3 writers is the most popular author 

### 2.4 Most Rated and Most Popular Books¶

* Most Rated Books: This is a list with the top 10 books with higher average ratings;
* Most Popular Books: This lists  the top 10 books with a higher total number of ratings.

```python
most_rated_books = df[['BookID', 'avg_rating']].sort_values(by='avg_rating', 
                                                        ascending=False).set_index('BookID').head(10)
most_popular_books = df[['BookID', 'num_of_ratings']].sort_values(by='num_of_ratings', 
                                                            ascending=False).set_index('BookID').head(10)
# using set_table_attributes to display two dataframes side by side
from IPython.display import display_html 
# creating a function to highlight Books ID in table
def highlight_cols(s):
    color = 'blue'
    return 'background-color: %s' % color
# applying styles and displaying
df1_styler = most_rated_books.reset_index().style.set_table_attributes(
                    "style='display:inline'").set_caption('Most Rated Books').applymap(highlight_cols, 
                    subset=pd.IndexSlice[:, ['BookID']]).format({'avg_rating': '{:,.1f}'})
df2_styler = most_popular_books.reset_index().style.set_table_attributes(
                        "style='display:inline'").set_caption('Most Popular Books').applymap(highlight_cols, 
                        subset=pd.IndexSlice[:, ['BookID']]).format({'num_of_ratings': '{:,.1f}'})
display_html(df1_styler._repr_html_() + ' ' + df2_styler._repr_html_(), raw=True)
```

The table above shows that the most rated books are not the most popular 

## 3 Most Recommended Books

### 3.1 Top 10 Most Recommended  Business Books 
Previously I calculated the Weighted_rating derived from (avg_rating * num_of_rating) this is to get the better selection of the recommended books. 
 
In this way, I can make a list of recommended books business books  considering the Avg Rating and the Number of Ratings each one received.
